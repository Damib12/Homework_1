```python
import geopandas as gpd

LinkPeru = "https://github.com/DACSS-CSSmeths-winter/spatialtest/raw/refs/heads/main/PeruMaps.gpkg"
peru = gpd.read_file(LinkPeru)
```

```python
# Filter to Lima only using the correct column name
lima = peru[peru["DEPARTAMEN"] == "LIMA"].copy()
```

```python
# Simplify geometry slightly to reduce file size
lima["geometry"] = lima["geometry"].simplify(tolerance=100)
```

```python
lima.to_file("lima.geojson", driver="GeoJSON")
```
